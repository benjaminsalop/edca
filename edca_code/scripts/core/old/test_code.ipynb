```python
# quick-diagnostics.py
from pathlib import Path
import pandas as pd
p = Path("summary_ranked_all.csv")
if p.exists():
    df = pd.read_csv(p)
    print("summary_ranked_all.csv shape:", df.shape)
    print("columns:", df.columns.tolist())
else:
    print("summary_ranked_all.csv not found in cwd")

# create a headless test plot
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
import numpy as np
out_dir = Path("outputs/figures")
out_dir.mkdir(parents=True, exist_ok=True)
x = np.linspace(0,10,100)
fig,ax = plt.subplots()
ax.plot(x, np.sin(x))
fig.savefig(out_dir/"diagnostic_test.png", bbox_inches="tight", dpi=150)
plt.close(fig)
print("Wrote diagnostic plot to", (out_dir/"diagnostic_test.png").resolve())
```
