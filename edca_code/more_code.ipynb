```python
import plotly.express as px
import pandas as pd

typology_df = pd.read_csv("/users/benjaminsalop/Desktop/Warneford_Comparison.csv")  # Replace with your actual path

# typology_df should be the cleaned dataframe
# Ensure last column is renamed:
typology_df = typology_df.rename(columns={typology_df.columns[-1]: "Total_kgCO2e_per_m2"})

fig = px.bar(
    typology_df,
    x="Typology",
    y="Total_kgCO2e_per_m2",
    title="Embodied Carbon per m² – Warneford Comparison",
)

fig.update_layout(
    xaxis_title="Structural Typology",
    yaxis_title="kgCO2e per m²",
)

fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd

# --- Data ---

data = [
    ["Solid Plank", 20.416, 12.4225, 0, 0, 0],
    ["Composite Deck", 50.348, 19.713, 31.5413, 0, 0],
    ["Hollowcore Slab", 42.5, 14.9307, 0, 0, 0],
    ["CLT Floor", 18, 0, 0, -168.4, 0],
]

df = pd.DataFrame(data, columns=[
    "Typology", "Concrete", "Rebar", "Steel", "Wood", "Others"
])

# Calculate totals
df["Total"] = df[["Concrete","Rebar","Steel","Wood","Others"]].sum(axis=1)

# Order typologies logically
order = ["Solid Plank", "Hollowcore Slab", "Composite Deck", "CLT Floor"]
df = df.set_index("Typology").loc[order].reset_index()

# --- Define material colours (consistent + clean) ---

material_colours = {
    "Concrete": "#8d8d8d",
    "Rebar": "#b5651d",
    "Steel": "#4f81bd",
    "Wood": "#2ca02c",
    "Others": "#9467bd"
}

# --- Create figure ---

fig = go.Figure()

materials = ["Concrete", "Rebar", "Steel", "Wood", "Others"]

for material in materials:
    fig.add_trace(
        go.Bar(
            name=material,
            x=df["Typology"],
            y=df[material],
            marker_color=material_colours[material],
            hovertemplate=
                "<b>%{x}</b><br>" +
                material + ": %{y:.2f} kgCO₂e/m²<extra></extra>"
        )
    )

# Overlay total as black marker line
fig.add_trace(
    go.Scatter(
        x=df["Typology"],
        y=df["Total"],
        mode="markers+lines",
        name="Total",
        marker=dict(color="black", size=10),
        line=dict(color="black", width=2),
        hovertemplate=
            "<b>%{x}</b><br>Total: %{y:.2f} kgCO₂e/m²<extra></extra>"
    )
)

fig.update_layout(
    barmode="relative",  # critical for negative timber
    title="Embodied Carbon per m² by Material and Structural Typology",
    xaxis_title="Structural Typology",
    yaxis_title="kgCO₂e per m²",
    legend_title="Material",
    template="plotly_white",
    height=600
)

fig.show()
```
